```python
courses = {"MCON 141" : "Intro to Programming",
           "MCON 232" : "Advanced Programming"}
print(courses)
# Add a new key value pair
courses["MCON 264"] = "Data Structures 1"

print(courses)

print(f"\n141: {courses["MCON 141"]}")

advanced = courses["MCON 232"]
print(advanced)

try:
  print(courses["MCON 641"])
except KeyError as key:
  print(f"Key not found: {key}")

if "MCON 232" in courses:
  print("MCON 232 found")
else:
  print("MCON 232 not found")

if "MCON 641" not in courses:
  print("Not found")
```
