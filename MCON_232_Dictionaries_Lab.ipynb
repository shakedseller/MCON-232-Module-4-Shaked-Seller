```python
# Create a dictionary, keys are numbers 1 - 5, values are the word "one" - "five"
num_dict = {1 : "one",
            2 : "two",
            3 : "three",
            4 : "four",
            5 : "five"}

even_dict = {key : value for key, value in num_dict.items() if key % 2 == 0}
print("num_dict:")
for key, value in num_dict.items():
  print(f"{key}, {value}")

print("\neven_dict:")
for key, value in even_dict.items():
  print(f"{key}, {value}")
```

## Assignment: Majority Element (Easy)
- Problem Given a list of integers nums, return the element that appears more than n/2 times.
- You may assume a majority element always exists.
- Example
- nums = [3, 2, 3] → 3
- nums = [2, 2, 1, 1, 1, 2, 2] → 2
- Goal Solve using a dictionary that maps: number -> count.



```python
def find_majority(nums):
  element_count = {}
  majority_elem = 0
  majority_count = 0

  for num in nums:
    element_count[num] = element_count.get(num, 0) + 1

  for key, value in element_count.items():
    if value > len(nums)/2:
      majority_elem = key
      majority_count = value

  #print(element_count)
  #print(majority_elem)

  return f"{majority_elem} appears {majority_count} times"

nums = [2, 2, 1, 1, 1, 2, 2]
find_majority(nums)
```

Leetcode 2 : Assignment: First Unique Character in a String Problem
Given a string s, return the index of the first non-repeating character.

If no such character exists, return -1.


```python
def unique(s):
  char_counts = {}

  for char in s:
    char_counts[char] = char_counts.get(char, 0) + 1

  target = ""
  for key, value in char_counts.items():
    if value == 1:
      target = key
      break

  index_unique = s.find(target)

  return index_unique

print(unique("Hello"))
print(unique("aaaae"))
```
